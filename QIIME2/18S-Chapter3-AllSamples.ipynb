Jupyter notebook with code for importing demultiplexed data from the 18S marker. For each library (9 total, a-i) 1) import 2) trim primers 3) summarize and view trimmed data 4) generate ESVs with quality trimming parameters based on (3) 5) summarize and view ESV table 6) tabulate and view ESV sequences
Then: 7) merge ESV table and sequences for all libraries 8) tabulate and view sequences, download as .fasta (GUI download) 9) summarize and view OTU table 10) export OTU table as biom 11) convert biom to tsv, json, and hdf formats

a: P1604 Library 1 b: P1604 Library 2 c: P1604 Library 3 d: BB Library 1 e: Hirai Library 1 f: Hirai Library 2 g: P2107 Library 1 h: ENP Library 1 i: GAK Library 1 

```python
#0: import the qiime2 python module to allow for inline visualizations
import qiime2 as q2
#which version of qiime2 and python are we using?
!qiime info
```

```python
#1.a import the fastq files for library 1 into a QIIME2 artifact
!qiime tools import \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-path /Users/stephaniematthews/Documents/Chapter3/qiime2-analyses/manifests_demultiplexed-data_into_qiime2/18S/P1604-Lib2-zhan_manifest_run1_grid.csv \
  --output-path a-paired-end-demux.qza \
  --input-format PairedEndFastqManifestPhred33
#2.a trim primers
!qiime cutadapt trim-paired \
    --i-demultiplexed-sequences a-paired-end-demux.qza \
    --p-front-f '^AGGGCAAKYCTGGTGCCAGC' \
    --p-front-r '^GRCGGTATCTRATCGYCTT' \
    --o-trimmed-sequences a-demux-pairedend-trimmed.qza \
    --verbose
#3.a summarize and view
!qiime demux summarize \
  --i-data a-demux-pairedend-trimmed.qza \
  --o-visualization a-demux-pairedend-trimmed.qzv
q2.Visualization.load('a-demux-pairedend-trimmed.qzv')
```

```python
#4.a Generate ESVs (Exact Sequence Variants). 
!qiime dada2 denoise-paired \
  --i-demultiplexed-seqs a-demux-pairedend-trimmed.qza \
  --p-trunc-len-f 273 \
  --p-trunc-len-r 216 \
  --p-pooling-method pseudo \
  --p-chimera-method pooled \
  --p-n-threads 0 \
  --p-n-reads-learn 10000 \
  --o-representative-sequences a-paired-end-demux_rep-seqs-dada2.qza \
  --o-table a-paired-end-demux_table-dada2.qza \
  --o-denoising-stats a-paired-end-demux_stats-dada2.qza
```
